## Necessary Functions

1.) Aşağıdaki fonksiyon gelen string değerlerini float() fonksiyonu ile string değerlerini float değerlere dönüştürür. İçeri verdiğimiz string değerlerinden borsa sitelerinde sayıları ve oranları ifade ederken kullanılan (% . ,) işaretlerini düzenledik.

```python
def string_to_float(string):
    if string is None:
        return np.nan
    if type(string)==float or type(string)==np.float64:
        return string
    if type(string)==int or type(string)==np.int64:
        return string
    str = string.split(" ")[0].replace('.','').replace(',','.').replace('%','');
    return float(str)
```

2.) String değerlerinden oluşan diziyi float değerlerinden oluşan diziye çevirir.

```python
def float_list(string_list):
    return list(map(string_to_float,string_list))
```

```python
3.) Aşağıdaki fonksiyon gelen string değerlerindeki birden fazla olan boşlukları kaldırır.
```

```python
def remove_multiple_spaces(string):
    if type(string)==str:
        return ' '.join(string.split())
    return string
```

## Making Http Requests in Python

```python
def get_table_simple(table,is_table_tag=True):
    elems = table.find_all('tr') if is_table_tag else get_children(table)
    table_data = list()
    for row in elems:
        row_data = list()
        row_elems = get_children(row)
        for elem in row_elems:
            text = elem.text.strip().replace("\n","")
            text = remove_multiple_spaces(text)
            if len(text)==0:
                continue
            row_data.append(text)
        table_data.append(row_data)
    return table_data
```

```python
def get_children(html_content):
    return [item for item in html_content.children if type(item)==bs4.element.Tag or len(str(item).replace("\n","").strip())>0]
```

```python
url = 'https://bigpara.hurriyet.com.tr/borsa/canli-borsa/'
tag, attr, value = 'table','id','markets_top' # Burada most unique outer tag girelim.

response = requests.get(url, timeout=2400)
content = BeautifulSoup(response.content, "html.parser")
content = content.find("div",attrs={'id':'sortable'})
# HTML(str(content))
result_list = get_table_simple(content, is_table_tag=False)
```

```python
df = pd.DataFrame(result_list)
df = df.set_index(0)
df
```
